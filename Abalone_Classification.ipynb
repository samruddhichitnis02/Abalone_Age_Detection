# Predict the age of abalone from physical measurements

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('abalone.data', header = None)
```

```python
data.head()
```

```python
data.shape
```

```python
data.columns = ['Sex', 'Length', 'Diameter', 'Height', 'Whole_weight', 'Shucked_weight','Viscera_weight',
               'Shell_weight','Rings']
```

```python
data.head()
```

```python
data['Rings'].unique()
```

# Rings  + 1.5  gives the age in years (given in description of data) 
'''Small Feature Engineering, Deriving Age from Rings Column, Age = Rings + 1.5'''

```python
data['Age'] = data['Rings']+1.5
```

```python
y = data['Rings'].values
```

```python
data.drop('Rings', axis = 1, inplace = True)
```

```python
data['Age'].unique()
```

```python
x = data.iloc[:,0:-1]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lb = LabelEncoder()
x.iloc[:,0:1] = lb.fit_transform(x.iloc[:, 0:1])
```

```python
x.head()
```

```python
ohe = OneHotEncoder(categorical_features = [0])
x = ohe.fit_transform(x).toarray()
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# Prediction Using LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
```

```python
score_lr = accuracy_score(y_test, y_pred_lr)
score_lr
```

# Prediction Using SVM

```python
from sklearn.svm import SVC
```

```python
sv = SVC(kernel='linear', C=1, gamma=1) 
sv.fit(x_train, y_train)
y_pred_sv = sv.predict(x_test)
```

```python
score_sv = accuracy_score(y_test, y_pred_sv)
score_sv
```

```python
sv = SVC(kernel = 'rbf', C = 1, gamma = 100)
sv.fit(x_train, y_train)
y_pred_sv = sv.predict(x_test)
```

```python
score_sv = accuracy_score(y_test, y_pred_sv)
score_sv
```

# Prediction Using LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
linear.fit(x_train, y_train)
y_pred_linear = linear.predict(x_test)
```

```python
from sklearn import metrics
mse = metrics.mean_squared_error(y_test, y_pred_linear)
rmse = np.sqrt(mse)
mse, rmse
```
